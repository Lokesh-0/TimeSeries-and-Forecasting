# Time Series Decomposition 

```python
result_add = seasonal_decompose(x=week['Weighted_Price'], model='additive', extrapolate_trend='freq', period=2)
plt.rcParams.update({'figure.figsize': (20,8)})
result_add.plot().suptitle('Additive Decompose', fontsize=25)
plt.show()
```

```python
result_add = seasonal_decompose(x=m['Weighted_Price'], model='multiplicative', extrapolate_trend='freq', period=2)
plt.rcParams.update({'figure.figsize': (20,8)})
result_add.plot().suptitle('multiplicative Decompose', fontsize=25)
plt.show()
```

# Rolling Mean and Standard Deviation 

```python
RollingMean = m['Weighted_Price'].rolling(window = 12).mean()
RollingSTD = m['Weighted_Price'].rolling(window = 12).std()
print(RollingMean)
print(RollingSTD)
```

```python
actual = plt.plot(m['Weighted_Price'], color = 'green', label='Actual')
mean = plt.plot(RollingMean, color = 'black', label='Mean')
std = plt.plot(RollingSTD, color = 'red', label='STD')
plt.legend(loc='best')
plt.title('Mean and STD')
plt.xlabel('Months')
plt.ylabel('Weighted Price')
plt.show()
```

# Dickey-Fuller Test

```python
from statsmodels.tsa.stattools import adfuller
test = adfuller(m['Weighted_Price'], autolag='AIC') 

result = pd.Series(test[0:4], index=['Dickey Fuller test Statistic:','P-Value:','Lags Used:','No. Of observations:'])
for key,value in test[4].items():
    result['Critical Value: %s' %key] = value
print(result)
```

We can see the P-Value is 0.993068, we cant reject the null hypothesis so the data is not stationary

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
from itertools import product
import warnings
```

```python
fig = plt.figure(figsize=(20,8))
ax1 = fig.add_subplot(211)
fig= sm.graphics.tsa.plot_acf(m['initial Diff'].iloc[13:],lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(m['initial Diff'].iloc[13:], lags = 50, ax=ax2)
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
Qs = range(0, 2)
qs = range(0, 4)
Ps = range(0, 4)
ps = range(0, 2)
D=1
d=1
parameters = product(ps, qs, Ps, Qs) #SARIMAX model Parameters:(p,d,q)(sp,sd,sq,s) +exog
parameterslist = list(parameters)
len(parameterslist)
results = []
bestaic = float("inf")
warnings.filterwarnings("ignore")
for param in parameterslist:
    try:
        model=sm.tsa.statespace.SARIMAX(m.Weighted_Price, order=(param[0], d, param[1]),seasonal_order=(param[2],D, param[3],12),enforce_stationarity=False,enforce_invertibility=False).fit(disp=-1)
    except ValueError:
        continue
    aic = model.aic
    if aic < bestaic:
        model1 = model
        aic1 = aic
        param1 = param
    results.append([param, model.aic])
```

```python
result = pd.DataFrame(results)
result.columns = ['parameters', 'aic']
print(result.sort_values(by = 'aic', ascending=True).head())
print(model1.summary())
```

```python
residual = model.resid
fig = plt.figure(figsize=(20,8))
ax1 = fig.add_subplot(211)
fig= sm.graphics.tsa.plot_acf(residual,lags=50, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(residual, lags = 50, ax=ax2)
```

```python
model.plot_diagnostics(figsize=(15, 12))
```

```python
fig = plt.figure(figsize=(20,8))
model1.resid.plot()
fig.suptitle('Residual Plot of the Best Model')
test1 = adfuller(model1.resid, autolag='AIC') 
result1 = pd.Series(test1[0:4], index=['Dickey Fuller test Statistic:','P-Value:','Lags Used:','No. Of observations:'])
print(result1)
```

```python
month_p = m[['Weighted_Price']]
upcoming = pd.DataFrame()
month_p = pd.concat([month_p,upcoming])
month_p['forecast'] = model1.predict(start=0, end=110)
plt.figure(figsize=(20,8))
plt.title('Bitcoin Predicted vs Actual')
plt.xlabel('Months')
plt.ylabel('USD')
month_p.Weighted_Price.plot()
month_p.forecast.plot(color='black', ls='-', label='Predicted Weighted_Price')
plt.show()
```
