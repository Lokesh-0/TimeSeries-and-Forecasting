```python
import requests
import pandas as pd 
from pandas.io.json import json_normalize
```

```python
url  = requests.get( "https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_DAILY&symbol=BTC&market=CNY&apikey=demo")
print(url.status_code)
#Parse into JSON
api_data = url.json()  

import pprint
pp = pprint.PrettyPrinter(indent=1)
pp.pprint(api_data)
```

```python
print(f"Request returned {url.status_code} : '{url.reason}'")
```

```python
#Loading from JSON to DataFrame
df = pd.DataFrame(api_data["Time Series (Digital Currency Daily)"]).T
df.columns = ['open CNY', 'open USD', 'high CNY', 'high USD',
   'low CNY', 'low USD', 'close CNY', 'close USD',
   'volume', 'market cap USD']
```

```python
df.info()
```

```python
df.reset_index(inplace=True)
df = df.rename(columns = {'index':'Date'})
```

```python
df = df.drop(['open CNY','high CNY','low CNY','close CNY','volume'], axis = 1)
```

```python
df
```

```python
df['open USD'] = df['open USD'].astype(float)
df['high USD'] = df['high USD'].astype(float)
df['low USD'] = df['low USD'].astype(float)
df['close USD'] = df['close USD'].astype(float)
df['market cap USD'] = df['market cap USD'].astype(float)
```

```python
import seaborn as sns
sns.lineplot(data=df["open USD"]) 
```

The Dataset is upside down since every new day's data in the API is adding to the first row

```python
#Reversing the Dataset
df1 = df.loc[::-1].reset_index(drop = True)
df1
```

```python
sns.lineplot(data=df1["open USD"])
```
